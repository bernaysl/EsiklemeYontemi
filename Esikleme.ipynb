```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
kitap_img = cv.imread('/content/kitapsayfasi.jpg')
kitap_img = cv.cvtColor(kitap_img, cv.COLOR_BGR2RGB)
```

```python
parlak_img = cv.convertScaleAbs(kitap_img,beta=100)
```

```python
plt.figure(figsize=(15,20))
plt.subplot(121), plt.imshow(kitap_img), plt.title('Orijinal')
plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(parlak_img), plt.title('Parlak')
plt.xticks([]), plt.yticks([])

plt.show()
```

fotograftaki renklerle isimiz yok, sadece kelimeleri gormek istiyorum. burada esikleme yontemi kullanacagiz.

```python
kitap_img = cv.cvtColor(kitap_img, cv.COLOR_RGB2GRAY)  #gri tonlamaya donusturuyorum.
```

```python
ret, thresh_img = cv.threshold(kitap_img,12,255, cv.THRESH_BINARY)  #esigimiz 12, eger sayi 12den buyukse 255 yap.
#ret=esik ve thresh=esikleme yapilmis goruntu
plt.figure(figsize=(12,8))
plt.imshow(thresh_img,cmap='gray')
```

```python
ret, thresh_img = cv.threshold(kitap_img,5,255, cv.THRESH_BINARY)  #esik degerini 5 yapalim ve karakterler kaybolsun

plt.figure(figsize=(12,8))
plt.imshow(thresh_img,cmap='gray')
```

```python
adaptive_thresh_img = cv.adaptiveThreshold(kitap_img,255,
                                           cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,15,2)  #ortalamasini kullanalim
                                                                                     #15'in yerine koydugumuz sayi her zaman tek sayi olmali
plt.figure(figsize=(12,8))
plt.imshow(adaptive_thresh_img, cmap='gray')
```
